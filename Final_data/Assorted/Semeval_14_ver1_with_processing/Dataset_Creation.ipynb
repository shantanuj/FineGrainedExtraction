```python
import xml.etree.ElementTree as ET 
import os
import csv 
import sys
import re
```

```python
xml_path = "./Final_data/Semeval_14_ver1/xmls/Laptops_Train.xml"
opinion_path = './Final_data/Semeval_14_ver1/Opinions/train_laptop'
output_csv_path = "./Final_data/Semeval_14_ver1/Combined_laptop.csv"
tree = ET.parse(xml_path) #xml dataset
with open(opinion_path,'r') as f: #opinions for dataset
    opinion_list = f.readlines()
root = tree.getroot()
output_list_inclusive = [["ID","Sentence","Aspects","Aspect_Polarities","Opinions","Opinion_Polarities"]]
output_list_separate = [["ID","Sentence","Aspect","Polarity","Opinion"]]
assert(len(root)==len(opinion_list))

for i in range(len(root)):
    
    sentence = root[i][0].text.replace(u'\xa0',u' ')
    sentence_id = root[i].attrib['id']
    opinions = opinion_list[i]
    opinions = re.sub(',','',opinions) #removing commas
    ops = re.split('([-+][01])?[\r]?[\n]?', re.sub(',','',opinions)) #splitting 
    opin_words = list(map(lambda x: x.strip(), ops[:-1:2])) #trimming opinion words
    if(opin_words == ['NIL']):
        opin_pols = ['NIL']
    else:
        try:
            opin_pols = list(map(lambda x: x.strip(), ops[1::2])) #opinion polarities
        except:
            print(i, ops)
    if(len(root[i])<=1):
        output_list_inclusive.append([sentence_id, sentence, "NIL","NIL",';'.join(opin_words),';'.join(opin_pols)])
    else:
        aspect_terms_list = []
        aspect_polarity_list = []
        for j in range(len(root[i][1])):
            aspect_terms_list.append(root[i][1][j].attrib['term'].replace(u'\xa0',u' '))
            aspect_polarity_list.append(root[i][1][j].attrib['polarity'])
        
        output_list_inclusive.append([sentence_id, str(sentence).strip(), ';'.join(aspect_terms_list),";".join(aspect_polarity_list),';'.join(opin_words),';'.join(opin_pols)])
    
with open(output_csv_path,'w') as f:
    wr = csv.writer(f)
    for row in output_list_inclusive:
        try:
            wr.writerow(row)
        except:
            print(row)
```
